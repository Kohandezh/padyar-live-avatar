```python
from TTS.config import load_config
from TTS.utils.manage import ModelManager
from TTS.utils.synthesizer import Synthesizer

model_path ="/content/best_model_male.pth" # Absolute path to the model checkpoint.pth
config_path ="/content/config.json" # Absolute path to the model config.json

synthesizer = Synthesizer(
        model_path,
        config_path,
        use_cuda=True

    )
```

```python

text="شرکت صنایع خودروسازی مدیران، که با نام مدیران خودرو نیز شناخته می‌شود، یک شرکت مونتاژ خودروی چینی است، که از سال۱۳۸۱ فعالیت خود را در منطقه ویژه اقتصادی اِرگ جدید در شهر بم آغاز کرد."

wavs = synthesizer.tts(text)
synthesizer.save_wav(wavs, 'sp.wav')
import IPython
IPython.display.Audio('sp.wav')
```
